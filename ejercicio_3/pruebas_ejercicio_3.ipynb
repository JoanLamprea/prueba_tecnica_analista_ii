```python
import sqlite3

conexion_db = sqlite3.connect('database.db')
cursor = conexion_db.cursor()
```

```python

#### Punto_3_1
with open('sql/01_clientes_sin_duplicados.sql', 'r') as sql_file:
    sql_01 = sql_file.read()

cursor.executescript(sql_01)
conexion_db.commit()

cursor.execute('SELECT tipo_doc, num_doc, cod_apli_prod, num_cta, f_vinc FROM clientes_sin_duplicados')
for row in cursor.fetchall():
    print(row)
```

```python
cursor.execute('DROP TABLE IF EXISTS transacciones_filtradas')
```

```python

#### Punto_3_2
with open('sql/02_transacciones_filtradas.sql', 'r') as sql_file:
    sql_02 = sql_file.read()

cursor.executescript(sql_02)
conexion_db.commit()

cursor.execute("""SELECT cod_trn FROM transacciones_filtradas""")
for row in cursor.fetchall():
    print(row)
```

```python
#### Punto_3_3
with open('sql/03_resultado_agrupado.sql', 'r') as sql_file:
    sql_03 = sql_file.read()

cursor.executescript(sql_03)
conexion_db.commit()


cursor.execute('''SELECT f_aplicacion_trn, num_doc, total_monto
FROM resultado_agrupado;
''')
for row in cursor.fetchall():
    print(row)
```

```python
def exportar_a_txt():
    column_names = ["Fecha de Aplicacion", "Numero de Documento", "Total Monto"]

    with open("resultado_agrupado.txt", "w") as archivo:
        archivo.write("\t".join(column_names) + "\n")
        for row in cursor.execute("SELECT f_aplicacion_trn, num_doc, total_monto FROM resultado_agrupado"):
            archivo.write("\t".join(map(str, row)) + "\n")
            
exportar_a_txt()
conexion_db.close()
```
